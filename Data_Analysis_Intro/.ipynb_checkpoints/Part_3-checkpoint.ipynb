```python
import pandas as pd

#Sometimes the coding is off - default should be "utf-8" - if that doesn't work try encoding "latin" 
df = pd.read_csv("Minimum Wage Data.csv", encoding="latin")

#Save a dataset into directory - 
df.to_csv("minwage.csv", encoding = "utf-8")
```

```python
df = pd.read_csv("minwage.csv")
df.head()
```

```python
#.groupby() will group columns in dataset by said "group"
gb = df.groupby("State")

#get an individual group
gb.get_group("Alabama").set_index("Year").head()
```

```python
#iterate over the group to get all the values of the group
act_min_wage = pd.DataFrame()

for name, group in df.groupby("State"):
    if act_min_wage.empty:
        act_min_wage = group.set_index("Year")[["State.Minimum.Wage.2020.Dollars"]].rename(columns={"State.Minimum.Wage.2020.Dollars":name})
    else:
        act_min_wage = act_min_wage.join(group.set_index("Year")[["State.Minimum.Wage.2020.Dollars"]].rename(columns={"State.Minimum.Wage.2020.Dollars":name}))
        
act_min_wage.tail()
```

```python
#figuring out issues with Alabama Data
issue_df = df[df["State.Minimum.Wage.2020.Dollars"]==0]
issue_df.head()
```

```python
#States Which We Cannot get information on
issue_df["State"].unique()
```

```python
#remove all the data which is useless
import numpy as np

#Note: axis default is 0, which is rows. We don't want to get rid of rows we want to get rid of columns.
act_min_wage.replace(0, np.NaN).dropna(axis=1).corr().head()
```

```python
#We don't know if ALL the information is bad with these states... Maybe in 1968 Texas didn't have a minimum wage... This could be true.
min_wage_corr = act_min_wage.replace(0, np.NaN).dropna(axis=1).corr()


for problem in issue_df['State'].unique():
    if problem in min_wage_corr.columns:
        print("we're missing something here...")
```

```python
#This is grouping up the State's, get getting Alabama's infor
grouped_issues = issue_df.groupby("State")
grouped_issues.get_group("Alabama").head()
```

```python
#let's sum up the State Min Wage over time - if it is greater than 1 THAN we know they must have recorded sometime some sort of data
grouped_issues.get_group("Alabama")['State.Minimum.Wage.2020.Dollars'].sum()
```

```python
#let's iterate over the groups and see which one gives us a value NOT equal to zero
for state, data in grouped_issues:
    if data['State.Minimum.Wage.2020.Dollars'].sum() != 0.0:
        print('we missed something...')
```
